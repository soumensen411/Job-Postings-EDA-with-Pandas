# Q4. Value Counts & Grouping What are the top 5 most common job_domain values? For each domain, calculate the average number of views and average number of applies.

```python

top_5_job_domains = df['job_domain'].value_counts().head().index.tolist() # the top 5 most common job_domain values
print("top 5 most comon job_domain are : ", top_5_job_domains)
```

```python

df_top_domains = df[df['job_domain'].isin(top_5_job_domains)]
avg_values = df_top_domains.groupby('job_domain')[['views','applies']].mean()
print(avg_values)
```

# Q5. String Cleaning The compensation column contains values like "$42,000.00 ". Convert it to a numeric column (float). Then find the median compensation for each pay_period type (YEARLY, HOURLY, MONTHLY).

```python
df['compensation']
```

```python
df['compensation'] = df['compensation'].str.replace(r'[$,]', '', regex=True).astype(float) # remove the $ symbol
```

```python
df['pay_period'].unique()
```

```python
df['compensation'].groupby(df['pay_period']).median()
```

# # Q6. Sorting & Ranking Find the top 10 job titles with the highest views. Among these, how many are sponsored vs. non-sponsored?

```python
top_views_index = df.sort_values(by='views',ascending=False).head(10)
top_views_index[['title','views']]  # the top 10 job titles with the highest views.
```

# Q7. GroupBy + Aggregation Group jobs by work_type and compute: 1. Total number of jobs 2. Mean applies 3. Max views

```python
df.groupby('work_type').agg({"job_id":"count","applies":"mean","views":'max'})
```

# Q8. Pivot Table Create a pivot table showing the average compensation for each combination of work_type (rows) and pay_period (columns). Which combination has the highest average compensation?

```python
df.pivot_table(index = 'work_type',
               columns='pay_period',
               values='compensation',
               aggfunc='mean'
               )
```

# Q9. Apply & Lambda Using the job_location column, extract the state abbreviation (e.g., "Albany, GA" → "GA"). Add it as a new column called state. Then find the top 3 states by number of job postings.

```python
df['state'] = df['job_location'].apply(
    lambda x : x.split(',')[-1].strip()  # extract the state abbreviation
)
```

```python
df['demand'] = df['applies'].apply(
    lambda x : "High" if x>100 else 'Low'
)
```

```python
df['state'].value_counts().head(3) # the top 3 states by number of job postings.
```

# Q10. Correlation & Insight Is there a correlation between views and applies? Calculate the Pearson correlation coefficient. Then create a new column conversion_rate = applies / views (handle division by zero). Which job_domain has the highest average conversion rate?

```python
df[['views','applies']].corr() # correlation between views and applies
```

```python
df['conversion_rate'] = df['applies'] / df['views'].replace(0, float('nan'))
df.groupby('job_domain')['conversion_rate'].mean().sort_values(ascending =False).head(1) # job_domain has the highest average conversion rate
```
